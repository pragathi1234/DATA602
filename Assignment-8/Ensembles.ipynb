### Fill target variables with zero, and drop unnecessary columns

```python
data['arr_delay'] = data['arr_delay'].fillna(0)
data.drop(columns=['manufacturer','model','tailnum','dest','carrier'],inplace=True)
```

```python
data.dtypes
```

```python
data.corr()
```

#### Split data into train and test

```python
X=data.drop(["arr_delay"],axis=1)
y=data["arr_delay"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print(f'Training examples: {X_train.shape[0]}')
print(f'Test examples: {X_test.shape[0]}')
```

### Import sklearn libraries

```python
from sklearn.model_selection import cross_val_score,GridSearchCV,cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

### Create pipeline

```python
cat_vars=['engine','origin','type']
num_vars=['temp','dewp','humid','wind_dir','wind_speed','wind_gust','precip','pressure','visib','air_time']

num_pipeline = Pipeline([('impute_missing', SimpleImputer(missing_values = np.nan,strategy='mean')),
                           ('standardize_num', StandardScaler())
                        ])

cat_pipeline = Pipeline([('impute_missing', SimpleImputer(missing_values = np.nan ,strategy='most_frequent')),
                          ('create_dummies', OneHotEncoder(handle_unknown='ignore',sparse=False))])


processing_pipeline = ColumnTransformer(transformers=[('proc_numeric', num_pipeline, num_vars),
                                                      ('create_dummies', cat_pipeline, cat_vars)
                                                     ])
print(processing_pipeline)
```

```python
p1 = Pipeline([('processing', processing_pipeline),
               ('LogisticRegression', LogisticRegression(penalty='l2', C=0.01, solver='lbfgs', random_state=1, max_iter=1000))])

p2 = Pipeline([('processing', processing_pipeline),
               ('DescisionTree', DecisionTreeClassifier(max_depth=2, criterion='entropy'))])

p3 = Pipeline([('processing', processing_pipeline),
               ('SVC', SVC(kernel='rbf',C=1, probability=True))])

labs = ['Logistic Regression', 'Decision Tree', 'SVM']
clfs = [p1, p2, p3]
clfs = zip(labs, clfs)

for lab, clf in clfs:
    scores = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=10, scoring='accuracy')
    print(f'Accuracy {scores.mean():.2%} [{lab}]')
```

```python
from sklearn.ensemble import VotingClassifier

ems = [('lr', p1),('dc', p2),('svc', p3)]
clf4 = VotingClassifier(estimators= ems, weights=None, voting='soft')

scores = cross_val_score(estimator=clf4, X=X_train, y=y_train, cv=10, scoring='accuracy')
print(f'Accuracy {scores.mean():.2%} [Ensemble]')
```

### Ensemble method didn't improve the model's accuracy.

```python
from sklearn.model_selection import GridSearchCV

params = {'lr__LogisticRegression__C':[0.001, 0.1, 1, 10], 
          'dc__DescisionTree__max_depth': [1,2,3],
          'svc__SVC__C':[0.001, 0.1, 1]
         }

vc_gs = GridSearchCV(estimator=clf4, param_grid=params, scoring='accuracy', refit=True, cv=5)
vc_gs = vc_gs.fit(X_train, y_train)
vc_gs_score = vc_gs.score(X_test, y_test)

print(f'VotingClassifier Test Accuracy: {vc_gs_score:.2%}')
```

```python
from sklearn.metrics import accuracy_score

tree = p2
tree = tree.fit(X_train, y_train)
tree_train_predict = tree.predict(X_train)
tree_test_predict = tree.predict(X_test)

tree_train_score = accuracy_score(y_train, tree_train_predict)
tree_test_score = accuracy_score(y_test, tree_test_predict)

print(f'Tree Training Score: {tree_train_score:.2%}')
print(f'Tree Test Score: {tree_test_score:.2%}')
```

```python
from sklearn.ensemble import AdaBoostClassifier

p4 = Pipeline([('processing', processing_pipeline),
        ( 'AdaBoost',AdaBoostClassifier())])
ada =p4

ada = ada.fit(X_train, y_train)
ada_train_predict = ada.predict(X_train)
ada_test_predict = ada.predict(X_test)

ada_train_score = accuracy_score(y_train, ada_train_predict)
ada_test_score = accuracy_score(y_test, ada_test_predict)

print(f'AdaBoost Training Score: {ada_train_score:.2%}')
print(f'AdaBoost Test Score: {ada_test_score:.2%}')
```

### By applying the boosting strategy, accuracy was increased by about 5%.
